```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('fruit.csv',sep='\t')
new_df = df.dropna()
#print(new_df.to_string())

x_axis_bar=df['fruit_name'].unique()
y_axis_bar=df['fruit_name'].value_counts(sort=False)

x_axis_scat=df['fruit_subtype'].unique()
y_axis_scat=df['fruit_subtype'].value_counts(sort=False)

x_axis_line=df['mass']
y_axis_line=df['mass']

piplot=df['fruit_name'].value_counts(sort=False)

explode=([.1,0,0,.1,])

plt.pie(piplot,shadow=True,labels=piplot,explode=explode)
plt.legend(title="fruits")
plt.figure(figsize=(15, 5))
plt.scatter(x_axis_scat,y_axis_scat)
plt.grid(color="black")
plt.show()
plt.figure(figsize=(10, 5))
plt.plot(x_axis_line,linestyle="dashdot",marker="o",mfc="green")
plt.grid(color="black")
plt.show()
plt.bar(x_axis_bar,y_axis_bar)
plt.grid(color="black")
plt.show()

```
